```python
#import libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Reading Data
df=pd.read_csv("http://bit.ly/w-data")
df.head()
```

```python
#data preprocessing
df.isnull().sum()
```

```python
# Plotting the distribution of scores
plt.scatter(df["Hours"],df["Scores"])
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.title("Study Hours vs Scores")
```

# Preparing data & Training Model

```python
x = df.iloc[:, :-1].values  
y = df.iloc[:, 1].values
```

```python
#Split data for training and testing
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
#Training Model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x_train,y_train)
print("Model Training Complate")
```

```python
print('Coefficients: ', reg.coef_)
print('Intercept: ',reg.intercept_)
```

```python
# Plotting the regression line
y_line=reg.coef_*x + reg.intercept_

plt.scatter(x,y)
plt.plot(x,y_line)
plt.show()
```

```python
#Predicte Data
y_pred=reg.predict(x_test)
print(pd.DataFrame({'Hours':[i[0] for i in x_test],'Precited Score':[j for j in y_pred]}))
```

```python
#Compare with Actual Data
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df1 
```

# Evaluating the model

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
```

### Predict Score if Student Study for 9.25 hours/Day

```python
hours=9.25
pred_score=reg.predict([[hours]])
print("Form {} hours of study, Student will Score: {}".format(hours,pred_score[0]))
```
